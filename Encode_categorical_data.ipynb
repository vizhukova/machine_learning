# One hot encoding
Кожна унікальна категорія перетворюється в бінарний вектор фіксованої довжини, де всі елементи рівні 0, за винятком одного, який вказує на належність до конкретної категорії. Цей метод дозволяє представити категоріальні дані у вигляді, зрозумілому для моделей.


```python
df_object_dummies = pd.get_dummies(my_object_df, drop_first = True)
print(df_object_dummies.shape)
df_object_dummies
```

```python
enc = OneHotEncoder() 
encoder = ce.OneHotEncoder(cols=my_object_df.columns, return_df=True, use_cat_names=True)
one_hot_encoded = encoder.fit_transform(my_object_df)
print(one_hot_encoded.shape)
one_hot_encoded
```

```python
# in both sets existing
intersection = one_hot_encoded.columns.intersection(df_object_dummies.columns)
print(len(intersection))
intersection
```

```python
# in both sets besides intersection
symmetric_difference = one_hot_encoded.columns.symmetric_difference(df_object_dummies.columns)
print(len(symmetric_difference))
symmetric_difference
```

```python
# only in one_hot_encoded set without data from intersection
difference = one_hot_encoded.columns.difference(df_object_dummies.columns)
print(len(difference))
difference
```

```python
# only in df_object_dummies set without data from intersection
difference = df_object_dummies.columns.difference(one_hot_encoded.columns)
print(len(difference))
difference
```

# Label encoding
Кожна унікальна категорія призначається унікальному цілочисельному значенню. Цей підхід використовується, коли категоріальні значення мають порядок або можна встановити певні взаємовідношення між ними.


```python
from sklearn.preprocessing import LabelEncoder
```

```python
def covert_to_number_categorical_numbers(column):
    return column.astype('category').cat.codes 

encoded = my_object_df.apply(covert_to_number_categorical_numbers, axis=0) 
encoded
```

```python
label_encoder = LabelEncoder() 
encoded = my_object_df.apply(lambda column: label_encoder.fit_transform(column), axis=0) 
encoded
```

```python
def label_encoder(df, categorical_columns=None):
    df = df.copy()
    """Encode categorical values as integers (0,1,2,3...) with pandas.factorize. """
    # if categorical_colunms are not given than treat object as categorical features
    if not categorical_columns:
        categorical_columns = [col for col in df.columns if df[col].dtype == 'object']
    for col in categorical_columns:
        df[col], uniques = pd.factorize(df[col]) # change categorical feature to numerical values
    return df, categorical_columns
label_encoder(my_object_df)
```

# Ordinal Encoding
Схожий на Label Encoding, але дозволяє вказати порядок між категоріями. Наприклад, якщо категорії мають визначений порядок, такий як "низький", "середній", "високий", цей порядок враховується при присвоєнні цілочисельних значень.


```python
sizes = ['small', 'medium', 'large']

# Encode colors as incremental integers 
encoded_sizes = np.arange(len(sizes))

print(encoded_sizes)


encoder = OrdinalEncoder()

# reshape to 2D array
sizes_reshaped = np.array(sizes).reshape(-1,1)

encoded = encoder.fit_transform(sizes_reshaped) 

print(encoded)
```
